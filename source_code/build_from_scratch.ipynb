```python
def normalize_data(data):
    if isinstance(data, pd.DataFrame):
        normalized_df = data.copy()
        for column in data.columns:
            max_value = data[column].max()
            normalized_df[column] = data[column] / max_value
        return normalized_df
    else:
        max_value = data.max()
        normalized_series = data / max_value
        return normalized_series
```

```python
data = pd.read_csv('Dataset_challenge.csv')

X_data = data.drop(columns=['Outcome'])  
y_data = data['Outcome'] 

X_data = normalize_data(X_data)

print(X_data)
```

```python
X_train = X_data.iloc[1:]
X_test = X_data.iloc[595:]
y_train = y_data.iloc[1:]
y_test = y_data.iloc[595:]
data = np.array(data)
m, n = data.shape
np.random.shuffle(data)
# Chuyển đổi DataFrames thành numpy arrays
X_train_np = X_train.to_numpy().T
X_test_np = X_test.to_numpy().T
y_train_np = y_train.to_numpy().T
y_test_np = y_test.to_numpy().T

# Kiểm tra kết quả
print(X_train_np)  
print(y_train_np) 
print(X_train_np.shape)
print(X_train.shape)
```

```python
def init_params():
    W1 = np.random.rand(10, 8) - 0.5
    b1 = np.random.rand(10, 1) - 0.5
    W2 = np.random.rand(2, 10) - 0.5
    b2 = np.random.rand(2, 1) - 0.5
    print(W1)
    print(W2)
    return W1, b1, W2, b2

def ReLU(Z):
    return np.maximum(Z, 0)

def softmax(Z):
    A = np.exp(Z) / np.sum(np.exp(Z), axis=0, keepdims=True)
    return A

def forward_prop(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1
    A1 = ReLU(Z1)
    Z2 = W2.dot(A1) + b2
    A2 = softmax(Z2)
    return Z1, A1, Z2, A2

def ReLU_deriv(Z):
    return Z > 0

def one_hot(Y, num_classes):
    one_hot_Y = np.zeros((num_classes, Y.size))
    one_hot_Y[Y, np.arange(Y.size)] = 1
    return one_hot_Y

def backward_prop(Z1, A1, Z2, A2, W2, X, Y):
    m = X.shape[0]  
    one_hot_Y = one_hot(Y, A2.shape[0])  

    dZ2 = A2 - one_hot_Y
    dW2 = dZ2.dot(A1.T) / m
    db2 = np.sum(dZ2) / m

    dZ1 = W2.T.dot(dZ2) * ReLU_deriv(Z1)
    dW1 = dZ1.dot(X.T) / m
    db1 = np.sum(dZ1) / m

    return dW1, db1, dW2, db2

def update_params(W1, b1, W2, b2, vW1, vb1, vW2, vb2, dW1, db1, dW2, db2, alpha, beta):
    vW1 = beta * vW1 + (1 - beta) * dW1
    vb1 = beta * vb1 + (1 - beta) * db1
    vW2 = beta * vW2 + (1 - beta) * dW2
    vb2 = beta * vb2 + (1 - beta) * db2

    W1 = W1 - alpha * vW1
    b1 = b1 - alpha * vb1
    W2 = W2 - alpha * vW2
    b2 = b2 - alpha * vb2

    return W1, b1, W2, b2, vW1, vb1, vW2, vb2
```

```python
def get_predictions(A2):
    return np.argmax(A2, axis=0)

def get_accuracy(predictions, Y):
    return np.sum(predictions == Y) / Y.size

# def SGD_with_momentum(X, Y, alpha, iterations, batch_size, beta=0.9):
#     W1, b1, W2, b2 = init_params()
#     vW1, vb1, vW2, vb2 = np.zeros_like(W1), np.zeros_like(b1), np.zeros_like(W2), np.zeros_like(b2)
#     n = X.shape[1]

#     for i in range(iterations):
#         permutation = np.random.permutation(n)
#         X_shuffled = X[:, permutation]
#         Y_shuffled = Y[permutation]

#         for j in range(0, n, batch_size):
#             X_batch = X_shuffled[:, j:j + batch_size]
#             Y_batch = Y_shuffled[j:j + batch_size]

#             Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X_batch)

#             dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W2, X_batch, Y_batch)

#             W1, b1, W2, b2, vW1, vb1, vW2, vb2 = update_params(
#                 W1, b1, W2, b2, vW1, vb1, vW2, vb2, dW1, db1, dW2, db2, alpha, beta
#             )

#         if i % 10 == 0:
#             predictions = get_predictions(A2)
#             accuracy = get_accuracy(predictions, Y_batch)
#             print(f"Iteration {i}, Accuracy: {accuracy:.4f}")

#     return W1, b1, W2, b2

def train_with_epoch(X, Y, alpha, epochs, batch_size, beta=0.9):
    W1, b1, W2, b2 = init_params()

    # Khởi tạo các giá trị động lượng (momentum) ban đầu bằng 0
    vW1, vb1, vW2, vb2 = np.zeros_like(W1), np.zeros_like(b1), np.zeros_like(W2), np.zeros_like(b2)
    n = X.shape[1]

    for epoch in range(epochs):
        # Xáo trộn dữ liệu ở mỗi epoch
        permutation = np.random.permutation(n)
        X_shuffled = X[:, permutation]
        Y_shuffled = Y[permutation]

        for j in range(0, n, batch_size):
            X_batch = X_shuffled[:, j:j + batch_size]
            Y_batch = Y_shuffled[j:j + batch_size]

            # Lan truyền xuôi
            Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X_batch)

            # Lan truyền ngược
            dW1, db1, dW2, db2 = backward_prop(Z1, A1, Z2, A2, W2, X_batch, Y_batch)

            # Cập nhật trọng số và bias với động lượng
            W1, b1, W2, b2, vW1, vb1, vW2, vb2 = update_params(
                W1, b1, W2, b2, vW1, vb1, vW2, vb2, dW1, db1, dW2, db2, alpha, beta
            )

        # In ra độ chính xác sau mỗi epoch
        if epoch % 1 == 0:
            Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)  # Tính dự đoán trên toàn bộ tập dữ liệu
            predictions = get_predictions(A2)
            accuracy = get_accuracy(predictions, Y)
            print(f"Epoch {epoch+1}/{epochs}, Accuracy: {accuracy:.4f}")

    return W1, b1, W2, b2
```

```python
# W1, b1, W2, b2 = SGD_with_momentum(X_train_np, y_train_np, 0.01, 500, 32)
W1, b1, W2, b2 = train_with_epoch(X_train_np, y_train_np, 0.01, 200, 32)
```

```python
def make_predictions(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    predictions = get_predictions(A2)
    return predictions

def test_prediction(index, W1, b1, W2, b2):
    current_data = X_train_np[:, index, None]
    prediction = make_predictions(X_train_np[:, index, None], W1, b1, W2, b2)
    label = y_train_np[index]
    print("Prediction: ", prediction)
    print("Label: ", label)
```
